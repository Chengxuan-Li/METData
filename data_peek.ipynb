```python
possible_vases = data[data['Classification'].isin(allowed_list)]
possible_vases.shape
```

```python
import requests
from bs4 import BeautifulSoup

def extract_featured_image(url: str):
    try:
        resp = requests.get(
            url,
            headers={'User-Agent': 'Mozilla/5.0'},
            timeout=5   # <-- add timeout here
        )
        soup = BeautifulSoup(resp.text, "html.parser")
        
        viewer = soup.select_one("div.image-viewer_wrapper__XTgZf")
        if not viewer:
            return None

        img = viewer.find("img", src=True)
        if not img:
            return None

        return img["src"]

    except Exception:
        # timeout / connection error / parsing error → return None
        return None



from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm

urls = possible_vases['Link Resource'].tolist()

with ThreadPoolExecutor(max_workers=10) as ex:
    results = list(tqdm(ex.map(extract_featured_image, urls),
                        total=len(urls)))


results
```

```python
with open("featured_images.txt", "w") as f:
    for r in results:
        f.write(f"{r}\n")
```

```python
possible_vases['image_link'] = results
```

```python
possible_vases_with_link = possible_vases[possible_vases['image_link'].notna()]
```
