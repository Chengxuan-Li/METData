```python
import requests
from bs4 import BeautifulSoup
import re

def extract_featured_image(url: str):
    resp = requests.get(url, headers={'User-Agent':'Mozilla/5.0'})
    soup = BeautifulSoup(resp.text, "html.parser")

    # find only the featured image section
    viewer = soup.select_one("div.image-viewer_wrapper__XTgZf")
    if not viewer:
        return None

    img = viewer.find("img", src=True)
    if not img:
        return None

    return img["src"]


from concurrent.futures import ThreadPoolExecutor
from tqdm import tqdm

urls = data['Link Resource'].tolist()

with ThreadPoolExecutor(max_workers=40) as ex:
    results = list(tqdm(ex.map(extract_featured_image, urls),
                        total=len(urls)))


results
```
